## The calculus of travel time

Synthetic seismograms can be created by doing basic calculus on travel-time functions. Integrating slowness (the reciprocal of velocity) can a yield a time-to-depth relationship for making a well tie. Differentiating an acoustic impedance (velocity times density) log yields a reflectivity function along the borehole, which can be convolved with a seismic wavelet. In effect, the integral tells us where a given interval of rock is in the time-domain, the derivative tells us what it should look like; its seismic character. 

My goal is to demonstrate the robustness of simple mathematical operations of travel time, 

The code and synthetic data required to reproduce the results and figures can be found here [ LINK ] in an accompanying IPython Notebook. For this tutorial, we will make use of the open-source Python libraries NumPy and SciPy, the later of which has a handy script for parsing well logs (LAS file)!

As an example, I will use the sonic and density log from the Penobscot L-30 well. This well sits within the same volume of 3D seismic data that Matt Hall used in his horizon smoothing tutorial, it's a vertical well, so we don't have to worry about deviated trajectories through time and space.

## Dealing with the shallow section

To deal with the shallow section, we first need to adjust the depths relative to sea level by subtracting the KB elevation (30.2 m) from the measured depths,

If we know integrate the sonic log, we see that time t = 0 corresponds to a depth of 347 m TVDss. To positioned the top of the log at the correct travel time on the seismic section, we need to use the thickness, and replacement velocities for both the water column, and the section above the log.

The replacement velocity for water = 1480 m/s

The replacement velocity for section above log = 1700 m/s

The units for this log are in feet (yuck). It turns out that we will have to convert to metres a number of times so let's write a function to do this,

```python
def f2m(item_in_feet):
    try:
        return item_in_feet / 3.28084
    except TypeError:
        return float(item_in_feet) / 3.28084
```

```python
print "KB elevation [m]: ", f2m(L30.well.KB.data) # Kelly Bushing (ft)
print "Seafloor elevation [m]: ", f2m(L30.well.GL.data) # Depth to Sea Floor (ft)
print "Top of sonic log [m]: ", f2m(L30.start)  # top of log (ft) (actually 1150 ft)

repl_int = f2m(L30.start) - f2m(L30.well.KB.data) + f2m(L30.well.GL.data)
water_vel = 1480
print "replacement interval [m]: ", repl_int

repl_vel = 1650 # m/s
repl_time = 2 * repl_int / repl_vel
print "two-way-replacement time: ", repl_time

water_time = 2 * np.abs(f2m(L30.well.GL.data)) / water_vel
print "seafloor bottom :", water_time

log_start_time = water_time + repl_time
print 'log_start_time:', log_start_time

def tvdss(md):
    "assumes a vertical well"
    return md - f2m(L30.well.KB.data)

top_log_TVDss = f2m(L30.well.KB.data) - f2m(L30.well.GL.data) 
```

### Fixing / editing the logs

We would like to clip spikes. But <code>np.clip()</code> clips absolute values. This won't work for us because there's a trend in the data. Instead, we'd like to compare to the neighbourhood, and clip locally spurious things. 

This is going to be a bit like, but not exactly like, the [conservative filter](http://subsurfwiki.org/wiki/Conservative_filter).

First, to make a statistic in a moving window, we need to use either convolution or make a loop. I'd like to use the median filter, which is nonlinear — so convolution is out. And loops are slow. So instead of resorting to a loop, we can simulate a rolling window by making shifted versions of the vector, then computing across the last dimension. This approach is from Erik Rigtorp, and involves making a new *n*+1 dimensional array that contains shifted copies of the array to roll across. Then we can compute the statistic we want across the new dimension. Cunning!

```python
def rolling_window(a, window):
        shape = a.shape[:-1] + (a.shape[-1] - window + 1, window)
        strides = a.strides + (a.strides[-1],)
        rolled = np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
        return rolled
```

Approach:

- Smooth with a rolling median filter (preserves edges, only attacks spikes). Note we are not actually going to use the smoothed version, but we are going to use the diff to decide where the spikes are.
- Find spikes by comparing the original data with the median-smoothed data. Use <code>np.where()</code> to list the indices meeting the clipping criterion.
- Replace the spikes with clipped versions of the curve: the local median plus (or minus) the clip value.

```python
def despike(curve, window, clip):
        
    # First we have to compute a smoothed version.
    curve_sm = np.median(rolling_window(curve, window), -1)
    curve_sm = np.pad(curve_sm, int(np.floor(window/2)), mode='edge')
    
    # Now we make a list of indices that differ from the local median
    # by at least the max_clip value.
    spikes = np.where(curve - curve_sm > clip)[0]  # +ve spikes
    spukes = np.where(curve_sm - curve > clip)[0]  # -ve spikes
    
    # Finally we replace those indices with a clipped version of the curve.
    out = np.copy(curve)
    out[spikes] = curve_sm[spikes] + clip  # Clip at the max allowed diff
    out[spukes] = curve_sm[spukes] - clip  # Clip at the min allowed diff
    
    return out
```

Now we know the indices of the spiky data, we can replace with a clipped version of the same spike. 

```python
rho = despike(curve=RHOB, window=13, clip=100)
```

### Computing the time-to-depth relationship

The time-to-depth relationship is obtained by scaling the sonic log by the sample interval (6" or 0.1524 m) and by calling <code>np.cumsum()<code> on it.

```python
tcum = 2 * np.cumsum(0.1524 * np.nan_to_num(dt) / 1000000.0 )
tdr = tcum + log_start_time
```

### Compute acoustic impedance

```python
I = (1e6/dt) * rho
```

```python
plt.figure(figsize=(18,4))
plt.plot(z,I)
plt.show()
```

### Compute reflection coefficient series

```python
RC = (I[1:] - I[:-1]) / (I[1:] + I[:-1])
```

```python
plt.figure(figsize=(18,4))
plt.plot(z[:-1],RC)
plt.show()
```

### Get well tops

Let's make a dictionary from the file 'tops.txt'

```python
tops = {}

with open('tops.txt') as f:
    for line in f.readlines():
        if not line.startswith('#'):
            temp = line.strip().split('\t')
            tops[temp[-1].replace('_',' ')] = {'depth':float(temp[1])}
```

Check we got what we wanted:

```python
tops
```

Make a function to accept a top find the nearest sample in time domain

```python
def find_nearest(array, value):
    idx = (np.abs(array - value)).argmin()
    return idx
```

For plotting tops in the time-domain, make a second entry in the dictionary for the travel-time.

```python
for top in tops:
    depth = tops[top]['depth']
    nearest_time = find_nearest(z, depth)
    tops[top]['time'] = tdr[nearest_time]
```

### Converting logs to two-way-travel time

But we need the functions represented by travel time

```python
# RESAMPLING FUNCTION
dt = 0.004
maxt = 3.0

t = np.arange(0, maxt, dt) 

Z_t = np.interp(x=t, xp=tdr, fp=Z)

RC_t = (Z_t[1:] - Z_t[:-1]) / (Z_t[1:] + Z_t[:-1])

plt.figure(figsize=(18,4))
plt.plot(z,Z)
plt.show()

plt.figure(figsize=(18,4))
plt.plot(t,Z_t)
plt.show()
```

## Creating the synthetic

Define a wavelet function — we will use a Ricker. 

```python
def ricker(f, length, dt):
    t = np.linspace(-length / 2, (length-dt) / 2, length / dt)
    y = (1. - 2.*(np.pi**2)*(f**2)*(t**2))*np.exp(-(np.pi**2)*(f**2)*(t**2))
    return t, y
```

Call this function, passing it the frequency *f*, the length in samples, and the sample interval d*t* in seconds. 

```python
tw, w = ricker (f=25, length=0.128, dt=0.004)
```

Perform a 1D convolution of the wavelet *w* with the reflection coefficient series <code>RC_t</code>:

```python
synth = np.convolve(w, RC_t, mode='same')
```
