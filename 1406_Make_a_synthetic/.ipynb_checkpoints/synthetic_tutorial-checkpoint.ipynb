## Despiking logs

We would like to clip spikes. But <code>np.clip()</code> clips absolute values. This won't work for us because there's a trend in the data. Instead, we'd like to compare to the neighbourhood, and clip locally spurious things. 

This is going to be a bit like, but not exactly like, the [conservative filter](http://subsurfwiki.org/wiki/Conservative_filter).

First, import libraries.

```python
import numpy as np
from las import LASReader
import matplotlib.pyplot as plt
%matplotlib inline
```

Read data as before.

```python
L30 = LASReader('L-30.las', null_subs=np.nan)
```

```python
z = L30.data['DEPTH']
RHOB = L30.data['RHOB']*1000
```

```python
plt.figure(figsize=(18,3))
plt.plot(z,RHOB)
plt.show()
```

Instead of resorting to a loop, we can simulate a rolling window by making shifted versions of the vector, then computing across the last dimension. This is from Erik Rigtorp, and Joe Kington [writes about it here](http://stackoverflow.com/questions/4936620/using-strides-for-an-efficient-moving-average-filter).

```python
def rolling_window(a, window):
        shape = a.shape[:-1] + (a.shape[-1] - window + 1, window)
        strides = a.strides + (a.strides[-1],)
        rolled = np.lib.stride_tricks.as_strided(a, shape=shape, strides=strides)
        return rolled
```

Now we can smooth with a rolling median filter (preserves edges, only attacks spikes). Note we are not actually going to use the smoothed version, but we are going to use the diff to decide where the spikes are.

```python
window = 13
rho_sm = np.median(rolling_window(RHOB,window), -1)
rho_sm = np.pad(rho_sm, window/2, mode='edge')
```

```python
plt.figure(figsize=(18,3))
plt.plot(z, rho_sm)
plt.show()
```

Check that the variance is reduced. 

```python
print np.amax(np.nan_to_num(RHOB)), np.amin(RHOB[np.where(~np.isnan(RHOB))])
print np.amax(np.nan_to_num(rho_sm)), np.amin(rho_sm[np.where(~np.isnan(rho_sm))])
```

Now we will find spikes by comparing the original data with the median-smoothed data. If they differ by more than 150 (arbitrary), then it's a spike. 

```python
spikes = np.where(RHOB - rho_sm > 150)[0]
spukes = np.where(rho_sm - RHOB > 150)[0]
```

Now we know the indices of the spiky data, we can replace with a clipped version of the same spike. 

```python
rho = np.copy(RHOB)
rho[spikes] = rho_sm[spikes] + 150  # Clip at the max allowed diff
rho[spukes] = rho_sm[spukes] - 150  # Clip at the min allowed diff
```

Let's examine a segment of the log to see what effect we had. 

```python
start = 14400
end = 15100

plt.figure(figsize=(18,4))
plt.plot(z[start:end], RHOB[start:end])
plt.plot(z[start:end], rho_sm[start:end])
plt.plot(z[start:end], rho[start:end])
plt.show()
```

And let's check the clipped version. 

```python
plt.figure(figsize=(18,3))
plt.plot(z, rho)
plt.show()
```
